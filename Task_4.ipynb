```python
categorical_columns = ['Gender','Degree', 'Specialization', 'CollegeState']

# Plotting frequency distribution for each categorical column
for column in categorical_columns:
    top_10_categories = df[column].value_counts().head(10)
    plt.figure(figsize=(10, 6))
    top_10_categories.plot(kind='barh')
    plt.title(f'Top 10 Frequency Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.xticks(rotation=45)
    plt.show()
```

```python
from scipy.stats import chi2_contingency
```

```python
contingency_table = pd.crosstab(df['Gender'], df['Specialization'])

chi2, p, dof, expected = chi2_contingency(contingency_table)
print("Chi-square statistic:", chi2)
print("P-value:", p)
```
